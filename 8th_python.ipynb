```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop
from keras.models import Sequential
from keras.layers import Activation
from keras.optimizers import SGD
from keras.layers import Dense
from keras.utils import to_categorical

import tensorflow as tf
import numpy as np
import time

(X_train, y_train), (X_test, y_test) = tf.keras.datasets.cifar10.load_data()
```

### Answer:
The new `model_2` has **1,352,510** parameters, which is **337,300 fewer** than the previous model (1,689,810 parameters). Because the first hidden layer was reduced from 500 neurons down to 400.

**Math: `model_2`:**
* **Input Layer:** 3072 features (32x32x3 images)
* **Layer 1 (Dense 400):** (3072 inputs × 400 neurons) + 400 biases = **1,229,200**
* **Layer 1 (Dropout 0.4):** **0** parameters
* **Layer 2 (Dense 300):** (400 inputs × 300 neurons) + 300 biases = **120,300**
* **Layer 2 (Dropout 0.4):** **0** parameters
* **Output Layer (Dense 10):** (300 inputs × 10 neurons) + 10 biases = **3,010**
* **Total = 1,352,510 parameters**

`part2:`

```python
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1] * X_train.shape[2] * X_train.shape[3])
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1] * X_test.shape[2] * X_test.shape[3])
y_train = y_train.ravel()
y_test = y_test.ravel()
```

```python
print("[INFO] training samples before scaling (3 pixels): {}".format(X_train[0][0:3]))

X_train = np.array(X_train) / 255.0
X_test = np.array(X_test) / 255.0
print("[INFO] training samples after scaling: {}".format(X_train[0][0:3]))
```

```python
classes = 10
print("[INFO] y0 before vectorization: {}".format(y_train[0]))

Y_train_vectors = to_categorical(y_train, classes)
y_test_vectors = to_categorical(y_test, classes)
print("[INFO] y0 after vectorization: {}".format(Y_train_vectors[0]))
```

```python
model_2 = Sequential()
#setting 1st hidden leyer to 400 and dropout to 0.4
model_2.add(Dense(400, input_dim=3072, activation="relu"))
model_2.add(Dropout(0.4))

#setting 2nd hidden Layer 300 and dropout to 0.4
model_2.add(Dense(300, activation="relu"))
model_2.add(Dropout(0.4))

#setting output layer to 10, cuz cifar has 10classes
model_2.add(Dense(10, activation='softmax'))

model_2.summary()
```

```python
print("[INFO] compiling model...")
#setting up RMSprop to 0.001
rmsprop_opt = RMSprop(learning_rate=0.001)

model_2.compile(loss="categorical_crossentropy", optimizer=rmsprop_opt, metrics=["accuracy"])

print("[INFO] training model_2...")
start = time.time()

model_2.fit(X_train, Y_train_vectors, epochs=20, batch_size=128, verbose=1)

print("[INFO] training time: {:.2f}s".format(time.time() - start))
```

```python
print("[INFO] evaluating model_2 on test set...")

(loss, accuracy) = model_2.evaluate(X_test, y_test_vectors, batch_size=128, verbose=1)

print("[INFO] loss={:.4f}, accuracy: {:.2f}%".format(loss, accuracy * 100))
```

```python
#given weightened inputs
z = np.array([0.4, 0.9, -0.5])

#sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

#softmax function
def softmax(x):
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x)

# 4. Calculate and print the results
print(f"Given Z values: {z}")

print("\n--- SIGMOID OUTPUTS ---")
sigmoid_outputs = sigmoid(z)
for i, val in enumerate(sigmoid_outputs):
    print(f"  Neuron {i+1} (z={z[i]}): {val:.4f}")

print("\n--- SOFTMAX OUTPUTS ---")
softmax_outputs = softmax(z)
for i, val in enumerate(softmax_outputs):
    print(f"  Neuron {i+1} (z={z[i]}): {val:.4f}")

print(f"\n[INFO] Total Softmax Sum: {np.sum(softmax_outputs):.1f}")
```
