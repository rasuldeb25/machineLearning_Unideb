```python
from sklearn.datasets import make_moons
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import warnings
```

**EXERCISE 1**

```python
X, y = make_moons(n_samples=500, noise=0.30, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
tree_numbers = range(5,301,15) # previous setting were: 15, 201, 10
oob_errors = []

warnings.filterwarnings('ignore', message='Some inputs do not have OOB scores') #we have this warning because we are starting from 5

#loopping through the range
for n_trees in tree_numbers:
    #initializing the Random Forest
    rf_model = RandomForestClassifier(n_estimators=n_trees, oob_score=True, random_state=42, n_jobs=-1)

    #fitting
    rf_model.fit(X_train, y_train)

    #error is being calculated as:  1 - accuracy
    oob_error = 1.0 - rf_model.oob_score_
    oob_errors.append(oob_error)

    print(f"  -> Trees: {n_trees:3d} | OOB Error: {oob_error * 100:.2f}%")
```

```python
plt.figure(figsize=(6, 4))

plt.plot(tree_numbers, oob_errors, marker='o', color='teal', linestyle='-', linewidth=2, markersize=6)

plt.xlabel("Num of Trees (n_estimators)", fontsize=12)
plt.ylabel("OOB Error", fontsize =12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
